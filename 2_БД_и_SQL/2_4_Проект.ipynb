### Приложение Finance Tracker

**Описание приложения**  
Приложение Finance Tracker будет помогать пользователям учитывать свои доходы и расходы. Основные функции:

- Добавление операций: Пользователь может добавлять доходы или расходы.
- Просмотр операций: Пользователь может просматривать список всех операций.
- Категории операций: Каждая операция относится к определенной категории (например, "Еда", "Транспорт", "Зарплата").
- Баланс: Приложение будет отображать текущий баланс пользователя.

**Структура базы данных**  
Мы создадим следующие таблицы:

- User: Пользователи приложения.
- Category: Категории операций (например, "Еда", "Транспорт").
- Transaction: Финансовые операции (доходы и расходы).
